```python
from global_vars import *
import numpy as np
import pandas as pd
import gc
import xarray as xr
import os
import glob
from datetime import datetime, timedelta
from tqdm import tqdm
from pathlib import Path
import fastparquet


input_vars = input_vars_ver1
output_vars_t = output_vars_ver1_dtcond
output_vars_q = output_vars_ver1_dcq


src_folder = os.path.join(data_folder, 'nodeep_train_data/')
src_files = glob.glob(os.path.join(src_folder, '*.csv'))

output_folder = os.path.join(data_folder, 'nodeep_train_data_ver1/')


dfs = [pd.read_csv(f) for f in src_files]
big_df = pd.concat(dfs, ignore_index=True).dropna()
big_df
```

```python
for var in big_df.columns:
    big_df[var] = pd.to_numeric(big_df[var], errors="coerce")


X = big_df[input_vars] ## use just temps for now
y_t = big_df[output_vars_t]
y_q = big_df[output_vars_q]


X.to_parquet(os.path.join(output_folder, "X_train.parquet"),engine="fastparquet")
y_t.to_parquet(os.path.join(output_folder, "y_t_train.parquet"),engine="fastparquet")
y_q.to_parquet(os.path.join(output_folder, "y_q_train.parquet"),engine="fastparquet")

gc.collect()
```
